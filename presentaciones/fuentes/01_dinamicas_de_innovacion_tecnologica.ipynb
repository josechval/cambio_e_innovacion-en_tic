# Definiciones preliminares.

## Tecnología.

**"*Grado de obtención del valor potencial de un recurso, mediante conocimientos y habilidades relativas al saber hacer y su combinación con recursos materiales, de manera sistemática, repetible y reproducible.*"**

-NMX-GT-001-IMNC-2007

## Innovación.

**"*Proceso dirigido a un mercado bajo un enfoque de negocio que detecta oportunidades y capacidades organizacionales para generar productos, procesos y servicios, novedosos aceptados por los consumidores.*"**

-NMX-GT-001-IMNC-2007

## Innovación tecnológica.

**"*Proceso que conjuga una oportunidad de mercado con una necesidad y/o una invención tecnológica, que tiene por objetivo la producción, comercialización y explotación de un nuevo proceso, producto, actividad comercial, modelo de negocio, modelo de logística o servicio al cliente.* "**

-NMX-GT-001-IMNC-2007

# Inteligencia de innovación.

La toma de decisiones no puede basarse en adivinación ni en "corazonadas", y aún cuando ciertos líderes cuentan con una gran sensibilidad e intuición, esto no es suficiente para justificar el despliegue de estrategias a largo plazo. 

Las decisiones en materia de innovación deben de estar respaldadas por datos que definan y delimiten aspectos críticos de la organización.

## Las iniciativas de innovación implican:

* Conocimiento del negocio.

* Conocimiento de los procesos.

* Conocimiento del mercado.

* Conocimiento tecnológico.

## Información para decidir.

En cualquier nivel, el uso de información veraz, completa y oportuna permite tomar decisiones razonadas.

Incluso con suficiente información histórica es posible realizar predicciones por medio de técnicas estadísticas y de análisis de datos.

## Datos -> Información -> Conocimiento.

La innovación tecnológica carecería de sentido si no se contara con información que la respaldara. Los datos, la información y el conocimiento son tema central de esta asignatura.

### El conocimiento de la organización.

La experiencia de las personas dentro de una organización es la más pura, pero a la vez la más escurridiza fuente de conocimientos. 

Es común que existan personas con conocimientos claves en las organizaciones, pero la fragilidad de la permanencia de este conocimiento es un constante riesgo que debe ser mitigado mediante la creación de bases de conocimiento y la comunicación. 

### El oceáno de datos.

En el otro extremo se encuentran los datos del negocio, de los procesos y de los mercados. Cualquier fenómeno o actividad que se pueda medir o incluso percibir es susceptible de convertirse en un dato y estos datos deben de ser procesados y analizados para poder ser útiles.

# Selección de proyectos de innovación.

Los procesos de innovación consumen diversos tipos de recursos tanto  de las personas y como de las organizaciones. 

Poder decidir (apostar) entre todas las opciones por aquellas que logren un equilibrio entre su impacto y su costo es tan importante como las ideas mismas.

## Innovación táctica.

Las organizaciones que no han adoptado a la innovación como parte de sus procesos se encuentran expuestas a un estado de inercia en el que la innovación sólo se da:

* Cuando una persona dentro de la organización descubre e impulsa una innovación (fenómeno Eureka).

* Cuando los factores de cambio obligan a la organización a evolucionar.

En este caso, la innovación es aleatoria y reactiva. 

No existe un proceso de innovación.

## Innovación estratégica.

Las organizaciones que incluyen a la innovación en sus procesos y organigrama son capaces de:

* Crear estructuras organizacionales que fomenten la innovación, tales como las áreas de investigación, desarrollo e innovación (I+D+I).

* Contar con información y conocimientos sobre su negocio, procesos y mercado que les permiten actuar de forma anticpada.

* Forjar alianzas con otros actores en condiciones favorables para aprovechar la innovación.

* Contar con un aservo de capital intelectual.

* Estar mejor preparadas para afrontar entornos adversos.

## Propiedad industrial.

La innovación es susceptible de tener un valor económico intrínseco.

Las ideas, los planes de negocios, la descripción de los procesos, las imágenes, los cursos y los documentos técnicos, aún cuando no tengan aplicación inmediata, pueden ser activos intangibles de las organizaciones.

# Comunicación con los implicados e interesados (stakeholders).

La capacidad de transmitir una idea a personas fuera del ámbito cultural y de interés de un individuo se vuelve un facilitador para la adopción de la innovación.

La comunicación se puede dar por medios formales tales como:
* Juntas de trabajo.
* Sesiones informativas.
* Reportes.
* Correos electrónicos.
* Conferencias y seminarios.

Sin embargo, en muchos casos los mensajes no llegan o no impactan.

Poder comunicar una idea de forma rápida y efectiva ante una persona con capacidad ejecutiva es una habilidad que puede definir el futuro de un negocio.
